# **Parametrlar va Argumentlar**

```python
def sum_of_two_numbers(a, b): #a va b - funksiyaning parametrlari
    s = a + b
    return s

sum_of_two_numbers(5, 7) #5 va 7 - funksiyaga berilgan argumentlar
```

```python
def salomlashuv(ism): #ism - parametr
    return f"Salom, {ism}"


salomlashuv(ism = "Hasan")
#quyidagicha yozsak ham bir xil bo'ladi
salomlashuv("Hasan")

#parametrga qiymat berdik
#Hasan - argument
```

# **Default arguments**

```python
#b parametrga argument berilmasa, qiymati 2 ga teng bo'ladi
#2 b parametrning default argumenti
def daraja_kotarish(a, b = 2):
    return pow(a, b)

print(daraja_kotarish(3)) #a = 3, b = 2
print(daraja_kotarish(2, 5)) # a = 2, b = 5
```

```python
def xayrlash(ism = "Mehmon"):
    return f'Xayr {ism}'

xayrlash() #argumentga Mehmon qiymatini oladi
xayrlash(ism = "Ravshan") #Ravshan so'zi argument sifatida berildi
```

# **args - arguments (argumentlar)**

funksiyadagi parametrlar soni aniq bo'lmaganda args kalit so'zidan foydalanamiz

```python
def ballni_tekshirish(*args):
    if sum(args) > 50:
        return "Barakalla o'g'lon!"
    else:
        return "O'qi ahmoq!"


ballni_tekshirish(2, 2, 3, 5, 5, 4)
```

# **kwargs - keyword arguments (kalit so'zli argumentlar)**

```python
def yigindini_hisobla(**kwargs):
    s = 0
    for key, value in kwargs.items():
        s += value
    return s

yigindini_hisobla(a = 4, b = 5, c = 7, d = 9)
yigindini_hisobla(e = 8, k = 6, m = 1, p = -5, i = 9)
```

# **DocStrings (Documentation String)**

```python
def sum_of_two_numbers(a: int, b: int):
    '''
    Bu funksiya ikkita sonning yig'indisini hisoblaydi

    Paramatrlar:
    a - float turidagi qiymat qabul qiladi
    b - float turidagi qiymat qabul qiladi

    Return:
    Berilgan ikkita sonning yig'indisi
    '''
    s = a + b
    return s

sum_of_two_numbers(a = 9.5, b = 4)
```

# **Parametrlar va funksiya qaytaradigan turni aniqlash**

```python
#Paramatrlar turini aniqladik
def salomlashuv(ism: str, familiya: str):
    return f"Salom, {ism} {familiya}"


salomlashuv("Ravshan", "Sodiqov")
```

```python
#funksiya qaytaradigan turni ham aniqladik
def salomlashuv(ism: str, familiya: str) -> str:
    return f'Salom, {ism} {familiya}'


salomlashuv("Hasan", "Hamidov")
```
