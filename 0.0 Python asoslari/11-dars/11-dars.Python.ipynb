# **Lambda funksiyalar**

Lambda funksiyalar kichik anonim funksiyalardir. Lambda funksiyasi har qanday miqdordagi argumentlarni qabul qilishi mumkin, lekin faqat bitta ifodaga ega bo'la oladi.

**`lambda argumentlar : ifoda`**

> **Yodda tuting:** *Lambda funksiyalarning nomi bo'lmaydi va faqatgina bitta ifodadan iborat bo'ladi*



## **Sonni kvadratga ko'taruvchi `square` funksiyasini yaratamiz**

```python
def square(number):
    return number ** 2

square(9)
```

## **Yuqoridagi funksiyani lambda funksiya tarzida yaratamiz**

```python
x = lambda number: number ** 2

x(8)
```

##**`map()` funksiyasi**

Ushbu funksiya list, tuple yoki shu kabi qator ma'lumotlarning har biri uchun ma'lum funksiyani bajarib, yangi map obyektini qaytaradi

```python
my_array = [3, 4, -1, 9, 0, 8]

list(map(lambda x: x ** 2, my_array))
```

##**`filter()` funksiyasi**

Ushbu funksiya list, tuple, yoki shu kabi qator ma'lumotlardan faqatgina ma'lum qoidaga mos keladiganlarini ajratib, yangi filter obyektini qaytaradi

```python
my_list = [10, 9.5, 20.4, 67.8, 16.9, 20, 22]

def saralash(number):
    if number > 15:
        return True
    else:
        return False

filtered_list = list(filter(saralash, my_list))

print(filtered_list)
```

Endi ushbu kodni lambda funksiya orqali qayta yozamiz

```python
my_list = [10, 9.5, 20.4, 67.8, 16.9, 20, 22]

filtered_list = list(filter(lambda number: number > 15, my_list))

print(filtered_list)
```
