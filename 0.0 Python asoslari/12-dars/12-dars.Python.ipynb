##`enumerate()`

```python
my_list = ['Ravshan', 'Bobomurod', 'Gulmira', 'Akhadjon', 'Hasan']

for raqam, element in enumerate(my_list):
    print(raqam, element)
```

# **Generatorlar**

```python
my_array = [3, 4, 5]

g = [x**2 for x in my_array]

type(g)
```

```python
my_array = [3, 4, 5]

g = (x**2 for x in my_array)

type(g)
```

##`yield`

```python
def counter(start, end):
    x = ''
    for i in range(start, end + 1):
        x = x + str(i) + " "
    return x

counter(7, 15)
```

```python
def counter(start, end):
    x = start
    while not x > end:
        yield x
        x += 1

list(counter(7, 15))
```
