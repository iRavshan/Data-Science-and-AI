```python
class Person:
    def __init__(self, name, age, gender, occupation, hobby):
        self.name = name
        self.age = age
        self.gender = gender
        self.occupatin = occupation
        self.hobby = hobby

    def say_hi(self):
        print(f'Salom, mening ismim {self.name}')

    def get_age(self):
        print(f'Salom, mening yoshim {self.age}')

    def get_gender(self):
        print(f'Men {self.gender}man')

    def say_your_hobby(self):
        print(f'My hobby is {self.hobby}')

    def get_occupation(self):
        print(f'My occupation is {self.occupatin}')
```

```python
person1 = Person(name="Ravshan", age=22, gender='male', occupation='teacher', hobby='football')
```
