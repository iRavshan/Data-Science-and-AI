# **Dunder metodlar**

`Dunder (double underscores)` funksiyalar ba'zida `magic` funksiyalar deb ham nomlanadi.

```python
class Person:
    #initialize
    def __init__(self, name, age, hobby):
        self.name = name
        self.age = age
        self.hobby = hobby

    #string
    def __str__(ravshan):
        return f'Person {ravshan.name}, yosh={ravshan.age}'

    #taqdim etmoq
    def __repr__(ravshan):
        return f'Person {ravshan.name}, yosh={ravshan.age}'
```

```python
dir(Person)
```

```python
person1 = Person(name="Ravshan", age=22, hobby="football")
```

```python
person1
#Person: Ravshan, yosh=22
```

```python
person2 = Person(name='Bobomurod', age=99, hobby='solving problems')
```

```python
class UFO:
    #initialize
    def __init__(self, first_name, last_name, age, hobby):
        self.full_name = first_name + " " + last_name
        self.age = age
        self.hobby = hobby
```

```python
ufo1 = UFO(first_name='John', last_name='Wick', age=1490, hobby='eating human being')
```

```python
ufo1.first_name
```

```python
ufo1.full_name
```

```python
from math import *

n = int(input())
s = 0

for i in range(1, n + 1):
    s *= sin(i) / pow(2, i)

print(f'{s:.2f}')
```

```python
import math

a, b, c, d = list(map(int, input().split()))

s = 0
p = 1
sp = 0
sp_sub = 1

for m in range(1, a+1):
    s += (3 * m**3  + 4*m + 5) / (m**3 + math.log(m))

for k in range(1, b+1):
    p *= k / (k**3 + 7*k + 5)


for i in range(1, c+1):
    for m in range(1, d+1):
        sp_sub *= (math.log(i) + math.pow(m, i)) / math.pow(m, i)
    sp += sp_sub
    sp_sub = 1

print(f'{s:.2f}', f'{p:.2f}', f'{sp:.2f}')
```
