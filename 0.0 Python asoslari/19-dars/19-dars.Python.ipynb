### **Sinflarda polimorfizm**

```python
class Animal:
    def __init__(self, name) -> None:
        self.name = name

    def info(self):
        return 'Salom, men hayvonman. Sizchi?!'
```

```python
class Person:
    def __init__(self, age) -> None:
        self.age = age

    def info(self):
        return 'Salom, men insonman, ahvollaring yaxshimi?!'
```

```python
animal = Animal(name='rex')
```

```python
animal.info()
```

```python
person = Person(age=40)
```

```python
person.info()
```

### **Vorislikda polimorfizm**

```python
class Fish:
    def __init__(self, name) -> None:
        self.name = name

    def swim(self):
        return 'Yahuu, men suzyabman!!!'
```

```python
class Shark(Fish):
    def swim(self):
        return 'Vaxaxa men akulaman!!!'
```

```python
shark = Shark(name='akulacha')
```

```python
shark.swim()
```
