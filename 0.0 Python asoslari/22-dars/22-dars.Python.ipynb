### **try ... except**

```python
try:
    a, b = list(map(float, input().split()))
    print(a / b)
except:
    print('Nimadir xato ketdi')
```

### **... finally**

```python
try:
    a, b = list(map(float, input().split()))
    print(a / b)
except:
    print('Nimadir xato ketdi')
finally:
    print('Bu finally ishlashi natijasida chiqdi')
```

### **`raise` kalit so'zi**

```python
try:
    a, b = list(map(float, input().split()))
    print(a / b)

except ValueError:
    print('Siz satr kiritdingiz')

except ZeroDivisionError:
    print("0 ga bo'lish amali bajarilmaydi")

except NameError:
    print("Dasturchining o'zi xato qildi")

except:
    print('Nimadir xato ketdi')
```

```python
def bolinma(number1, number2):
    try:
        return number1 / number2
    except:
        print('Uzr aka')

bolinma("Ravshan", 'qarshi')
```
