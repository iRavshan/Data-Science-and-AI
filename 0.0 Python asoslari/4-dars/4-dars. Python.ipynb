# **Satrlarni kesib olish**

Deylik sizga qaysidir satrning qandaydir qismigini kerak xolos. Endi siz bu satrning sizga kerakli qismini kesib olishingizga to'g'ri keladi.

Satrlarning qismini kesib olishni Python da juda ham oson bajarish mumkin. Bunda sizga faqatgina bitta formulaning o'zi kifoya:

**[*start* : *stop* : *step*]**

Deylik sizda qiymati **"Bugun havo juda issiq"** qiymatiga ega bo'lgan satr bor va sizga bu satrning **"juda issiq"** degan qismi kerak. Endi bunda:

*   j harfining turgan o'rni = 11
*   q harfining turgan o'rni = 20

> **Unutmang!** Bo'sh joy ham bitta o'rinni egallaydi.

va yana shuni doimo yodda tuting-ki,


> siz 20-o'rinda turgan elementgacha qirqib olmoqchi bo'lsangiz **stop uchun 20 emas 21 qiymat berishingiz kerak.** Chunki 21 qiymat berganingizda indeksi 21 gacha bo'lgan elementlar qirqib olinadi.

````python
#avvalo shu o'zgaruvchini yaratamiz
ob_havo = "Toshkent bizning poytaxt"

#kerakli qismni qirqib olamiz
print(ob_havo[0:8:3])
````

Deylik endi sizga ushbu satrning **"juda issiq"** qismi emas **"jd si"** qismi kerak. Ya'ni **"juda issiq"** degan qismni butunlay emas oradan har ikkinchi harfini qirqib olmoqchimiz.

Endi biz ushbu holatda **step qiymatiga 2 berishimiz kerak**. Agar sizda har 3-harf kerak bo'lsa 3 qiymat berasiz, 4-harf kerak bo'lsa 4 qiymat berasiz va hkz.

````python
#"jd si" qismini qirqib olamiz
print(ob_havo[11:21:2])
````

# **Satrlarni kesib olishda ayrim qisqartirib yozishlar**

````python
name = "Mening ismim Ravshan"

#start qiymatni tashlab yozamiz
qirqilgan_soz = name[:6:] #start = 0, stop = 6, step = 1
print(qirqilgan_soz)
````

````python
#start qiymatni tashlab yozamiz
qirqilgan_soz = name[:6:] #start = 0, stop = 6, step = 1
print(qirqilgan_soz)
````

````python
#stop qiymatni tashlab yozamiz
qirqilgan_soz = name[3::2] #start = 3, stop = oxirigacha, step = 2
print(qirqilgan_soz)
````

````python
#starni ham, stopni ham, stepni ham tashlab yozamiz
qirqilgan_soz = name[::]
print(qirqilgan_soz)
````

# **Ayrim string funksiyalari**

```lower() ``` - funksiyasi satrning barcha harflarini kichik harfga o'tkazib yangi satr qaytaradi.



````python
satr = "Azamat AJOYIB Yigit"
print(satr.lower())
````

```upper() ``` -  satrning barcha harflarini katta harfga o'tkazib yangi satr qaytaradi.

````python
satr = "U yaqinda Germaniyaga uchib ketmoqchi"
print(satr.upper())
````

```title() ``` -  satrdagi barcha so'zlarning birinchi harflarini katta harfga o'tkazib yangi satr qaytaradi.

````python
satr = "Jeff Bezos oz lavozimidan ketdi"
print(satr.title())
````

```swapcase() ``` -  satrdagi katta harflarni kichik harfga, kichik harfdagilarni kattalariga o'tkazib yangi satr qaytaradi.

````python
satr = "Jeff Bezos o'z lavozimidan ketdi"
print(satr.swapcase())
````

```isdigit() ``` -  satrdagi barcha elementlar raqamlardan iborat bo'lsa True, aks holda False qiymat qaytaradi

````python
satr = "12000"
print(satr.isnumeric()) #True qaytaradi

satr = "12000a"
print(satr.isdigit()) #False qaytaradi, chunki satrda a bor
````

```count() ``` -  funksiyaga qiymat sifatida berilgan elementning satrda necha marta bor ekanligini qaytaradi

````python
satr = "Anvar Ajoyib dasturchi"

#satrda a necha marta qatnashganini hisoblaymiz
print(satr.count('a'))
````
