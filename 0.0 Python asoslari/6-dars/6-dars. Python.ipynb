# **Ro'yxat elementlarini o'zgartirish**

Ro'yxatning faqatgina bitta elementini o'zgartirish

```python
a = [2, 7, 15, 8, 1, 6, 10, 14]

a[-6] = 'Qarshi'

a
```

Ro'yxatning bir nechta elementlarini o'zgartirish

```python
a = [2, 7, 15, 8, 1, 6, 10, 14]

a[0:7] = [1, 2, 'Toshkent', 4, 5.7, 6, 7]

a
```

# **Yangi elementlar qo'shish**

`insert()` - ro'yxatning ma'lum o'rniga yangi element qo'shish

```python
my_list = [9, 8, -1, 6.5, 'Qarshi']

my_list.insert(2, 'Toshkent')

my_list
```

`append()` - ro'yxatning oxiriga yangi element qo'shish

```python
my_list = [9, 8, -1, 6.5, 'Qarshi']

my_list.append('Toshkent')

my_list
```

`extend()` - ro'yxatning elementlarini ikkinchisi orqali kengaytirish

```python
mevalar = ['olcha', 'anor', 'uzum']

sonlar = [1, 2, 3, 4, 5]

mevalar.extend(sonlar)

mevalar
```

# **Ro'yxatdan elementlarni o'chirish**

`remove()` - berilgan qiymatdagi elementni o'chiradi

```python
aralash = ['olcha', 'anor', 'uzum', 1, 2, 3, 4, 5]

aralash.remove('uzum')

aralash
```

`pop()` - berilgan indeksdagi elementni o'chiradi

```python
aralash = ['olcha', 'anor', 'uzum', 1, 2, 3, 4, 5]

aralash.pop(2) #2-indeksdagi son o'chiriladi

aralash
```

`del` - berilgan indeksdagi elementni yoki ro'yxatni to'liqligicha o'chiradi

```python
del aralash

aralash
```

`clear()` - ro'yxatning barcha elementlarini o'chiradi ammo ro'yxatning o'zini o'chirmaydi

```python
numbers = [2, 4, 7.8, -9, -3,4]

numbers.clear()

numbers
```

# **Ro'yxat elementlarini saralash**

`sort()` - ro'yxat elementlarini saralash uchun qo'llaniladi

O'sish tartibida

```python
fruits = ['olcha', 'anor', 'uzum', 'banan']

fruits.sort(reverse=False)

fruits
```

Kamayish tartibida

```python
numbers = [2, 4, -1, 9.2, 16, -3, 1]

numbers.sort(reverse=True)

numbers
```

`reverse()` - elementlarni teskari tartibda joylashtirish
