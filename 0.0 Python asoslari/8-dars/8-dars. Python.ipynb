`if` - agar - bilan tanishamiz

```python
a = int(input())
b = int(input())

if a > b:
    print(a)
```

`else` - aks holda -  bilan tanishamiz

```python
a = int(input())
b = int(input())

if a > b:
    print(a)
else:
    print(b)
```

# **Ichma-ich shartlar**


image.png

`elif` - `else + if` - bilan tanishamiz

```python
x = 10
y = 9

if x == y:
    print('equal')
elif x < y:
    print('less')
else:
    print('greater')
```

`if else` ni bir qatorda yozish

```python
name = "Hasan"

print("Sizning ismingiz Hasan") if(name == "Hasan") else print("Siz boshqasiz")
```
