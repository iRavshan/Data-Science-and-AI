# **While loop**

```python
x = 20

while x > 15:
    #sikl tanasi
    x = x - 1 #19, 18, 17, 16, 15
    print("Ishlayapman!")

print(x)
```

`while + else` - while sikl oxirida `else` dan foydalanish

```python
x = 20

while x > 15:
    x = x - 1 #19, 18, 17, 16, 15
    print(x)
else:
    print("Bu sikldan keyin ishlaydigan kod")

print(x)
```

# **Looping through lists**

### **1-usul**

```python
my_list = [3, "Qarshi", "Abdukarim", 3+4j, 4.5, 'k']

for i in my_list:
    print(i)
```

### **2-usul**

```python
my_list = [3, "Qarshi", "Abdukarim", 3+4j, 4.5, 'k']

for i in range(len(my_list)):
    print(my_list[i])
```

# **Looping through strings**

```python
x = 'Ravshan bugun maktabga bormadi'

a = x.split()
#a = ['Ravshan', 'bugun', 'maktabga', 'bormadi']

for i in a:
    print(i)

#Input: Toshkent mening poytaxtim
'''
OUTPUT:
Toshkent
mening
poytaxtim
'''

#Input: Ravshan bugun maktabga bormadi
'''
OUTPUT:
Ravshan
bugun
maktabga
bormadi
'''
```

# **Nested Loops (Ichma-ich sikllar)**

```python
for i in range(6):
    print(i)
    for j in range(4):
        print(j)
        for k in range(3):
            print(k)
```

# **Break keyword**

```python
for i in range(10):
    if i == 4:
        break
    print(i)
```

berilgan ro'yxatdan birinchi uchragan ikkiga qoldiqsiz bo'linadigan songacha bo'lgan qiymatlar chop etilsin

```python
a = [7, 35, 7, 5, 2, 5, 4]

for i in a:
    if i % 2 != 0:
        print(i)
    else:
        break
```

# **Continue keyword**

```python
a = [3, 4, 5, 6, 7, 8, 9]

for i in a:
    if i == 6:
        continue
    print(i, end=' ')
```

Masala: Berilgan ro'yxatdan ikkiga qoldiqsiz bo'linadigan elementlaridan tashqari barcha elementlar chop etilsin

```python
a = [10, 14, 21, 23, 35, 9, 6]

for i in a:
    if i % 2 == 0:
        continue
    print(i, end=' ')
```
