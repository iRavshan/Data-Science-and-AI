# **Ro'yxatlarni NumPy massivlariga aylantirish**

Python listlarni NumPy massivlariga aylantirish

```python
#oddiy list yaratamiz
my_list = [2, 3, 10, 8.2, 3+2j]

#listni numpy massiviga aylantiramiz
array = np.array(my_list)

print(array)
```

Nested (ichma-ich) listlarni NumPy massiviga aylantirish

```python
#nested list yaratamiz
my_list = [[3, 4, 5], [8, 9, 3], [3, 4, 5]]

#listni numpy massiviga aylantiramiz
array = np.array(my_list)

print(array)
```

# **NumPy massivlar yaratish**

`arange()` - range() funksiyasining start, stop, step qoidasi kabi ishlaydi - **arange(start, stop, step)**

```python
np.arange(1, 10, 2)
```

`ones()` - faqatgina birlardan iborat mxn o'lchamli massiv yaratadi - **ones((m, n))**

```python
np.ones((2, 3))
```

`zeros()` - faqatgina nollardan iborat mxn o'lchamli massiv yaratadi - **zeros((m, n))**

```python
np.zeros((3, 4))
```

`eye()` - mxn o'lchamli massiv yaratadi - **eye(m, n, k)**

```python
np.eye(4, 4, k = 0)
```

`identity()` - nxn o'lchamli birlik matritsa yaratadi - **identity(n)**

```python
np.identity(4)
```

`linspace()` - - **linspace(start, stop, num)**

```python
np.linspace(10, 20) #axis = 0, star-vektor
```

# **NumPy massivlarning ayrim xususiyatlarini aniqlash**

```python
my_array = np.array([3, 4, 5, 9, 10, 12], dtype = np.int64)

my_array
```

`dtype` - massiv qiymatlarining turini aniqlash

```python
my_array.dtype
```

`shape` - massiv qanday mxn o'lchamda (shaklda) ekanligini aniqlash

```python
my_array.shape
```

`ndim` - massivning nechi o'lchamli ekanligini aniqlash

```python
my_array.ndim
```

`size` - massivning nechta elementga ega ekanligini (hajmini) aniqlash

```python
my_array.size
```

`itemsize` - massivning bitta elementi xotiradan qancha joy egallashini baytlarda aniqlash

```python
my_array.itemsize
```

`nbytes` - massiv xotiradan qancha joy egallashini baytlarda aniqlash

```python
my_array.nbytes
```
