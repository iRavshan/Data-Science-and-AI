# **Boolean indekslash**


Ma'lum shartga bo'ysunadigan elementlarga murojaat qilish

```python
arr = np.array([[0, 9], [-2, 4], [5, -6]])

arr[arr > 3]
```

Agar shart uzunroq bo'lsa shartni maska sifatida alohida yaratib oling

```python
mask = ~(arr == 4) & (arr > 3)

arr[mask]
```

# **Kesish**

Yana o'sha **Start, Stop, Step** qoidasi

# **Bir o'lchamli (1-D) massivlarni kesish**

image.png

```python
arr = np.arange(1, 11)

arr[3:7] #start = 3, stop = 7, step = 1
```

# **Ikki o'lchamli (2-D) massivlarni kesish**

image.png

```python
arr = np.array([[2, 4, 7], [-4, 5, 6], [-4, 5, 9]])

arr
```

```python
arr[1:, 0:2]
```

# **Uch o'lchamli (3-D) massivlarni kesish**

image.png

```python
#uch o'lchamli massiv yaratamiz
my_array = np.array([
            [[10, 11, 12], [13, 14, 15], [16, 17, 18]],
            [[20, 21, 22], [23, 24, 25], [26, 27, 28]],
            [[30, 31, 32], [33, 34, 35], [36, 37, 38]]
])

my_array[:2][1:3][:2]
```
