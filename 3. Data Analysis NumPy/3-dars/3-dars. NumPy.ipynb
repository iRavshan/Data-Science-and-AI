# **Massivlarni o'zgartirish (transformation)**

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

arr.shape
```

`reshape()` - shakli o'zgartirilgan yangi massiv qaytaradi

```python
arr.reshape(6, 1)
```

`resize()` - massivning shaklini o'zgartiradi, yangi massiv qaytarmaydi

```python
arr.resize(3, 2)

arr
```

`flatten()` - tep tekis qilib tashlamoq - massiv qanday o'lchamda bo'lsa ham, bitta qatordan iborat bir o'lchamli yangi massiv qaytaradi

```python
#flatten(order = 'C') - order = {'C', 'K', 'A', 'K'} (optional)
arr.flatten()
```

`ravel()` - flatten() bilan bir xil ishlaydi

image.png

```python
np.ravel(arr)

arr.ravel()
```

`transpose()` - massivning o'qlarini almashtirib yangi massiv qaytaradi

```python
arr.transpose(1, 0)
```

`swapaxes()` - massivning ikkita o'qlarini almashtirib yangi massiv qaytaradi

```python
np.swapaxes(arr, 0, 1)
```

`flip()` - massivning shaklini saqlab, belgilangan oq boylab massiv elementlari tartibini ozgartiradi.

```python
np.flip(arr, 0)
```

`astype()` - mavjud massivning ma'lumotlar turini o'zgartirib, yangi massiv qaytaradi

```python
arr = np.arange(10, dtype=np.uint16)

arr.dtype
```
