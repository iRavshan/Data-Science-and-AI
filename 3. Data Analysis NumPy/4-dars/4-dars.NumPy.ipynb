# **Random funksiyalar**

```python
from numpy import random
import numpy as np
```

# **NumPy massivini yaratish**

`random.rand(m, n)` - mxn o'lchamdagi 0 va 1 oralig'idagi haqiqiy sonlardan tashkil topgan massiv yaratadi (0, 1)

```python
random.rand(3, 2)
```

`random.randint(low, high=None, size=None, dtype=int)` - qiymatlari [low, high) oraliqda bo'lgan berilgan o'lchamdagi massiv yaratadi

```python
random.randint(low=3, high=8, size=(4, 3))
```

`random.choice(a, p=None)` - yozib tushuntirish biroz murakkab

```python
x = random.choice([0, 3, 7, 9, 8, -1], p=[0.1, 0.1, 0.2, 0.2, 0.3, 0.1], size=(3, 2))

x
```

`random.shuffle()` - massiv elementlarini random tarzida almashtirib tashlaydi

```python
x = np.arange(20)

random.shuffle(x)

x
```
