# **Chiziqli algebra masalalarini ishlash: Linear algebra moduli**#

```python
from numpy import linalg
import numpy as np

array1 = np.array([3, 4, 6, 8, 9])
array2 = np.array([0, 1, -1, 2, 3])
```

Matritsalarni qo'shish

```python
array1 + array2
```

Matritsalarni ayirish

```python
array1 - array2
```

Matritsalarni ko'paytirish

```python
#array1 * array2 xato bu usulda ko'paytirish
```

Matritsani songa ko'paytirish

```python
np.dot(array1, array2)

array1.dot(array2)

#array1 * array2 - usulida ishlash xato
```

Matritsani transponirlash

```python
array1 = np.array([3, 4, 6, 8, 9, 2]).reshape(3, 2)

(array1.T).shape

array1.transpose()
```


Matritsa determinantini hisoblash

```python
array = np.array([3, 4, 6, 8, 9, 2, -1, 0, 0]).reshape(3, 3)

linalg.det(array)
```

Matritsani (butun) darajaga ko'tarish

```python
linalg.matrix_power(array, 5)
```

Matritsaga teskari matritsani (inversiyasini) topish

```python
linalg.inv(array)
```

Matritsaga qarama-qarshi matritsani topish

```python
array * -1
```
