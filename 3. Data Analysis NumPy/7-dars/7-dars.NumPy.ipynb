# **NumPydagi qo'shimcha funksiyalar: Bitiruv oqshomi. Phew**

```python
import numpy as np
```

`concatenate()` - massivlarni birlashtirish

```python
x = np.arange(0, 9)
y = np.arange(10, 15, 2)

z = np.concatenate([x, y])
```

`stack()` - massivlarni o'q bo'yicha birlashitirish, vstack, hstack, dstack()

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

np.stack([x, y], axis=1)
```

`array_split(n)` - massivni n ta massivga bo'lish

```python
a = np.array([3, 4, 5, 6, 4])

np.array_split(a, 7)
```

`where()` - massivdan kerakli elementlarning indekslarini qaytaradi

```python
a = np.array([3, 4, 5, 6, 4, 12, 3])

x = np.where(a % 3 == 0)

type(x)
```

`searchsorted()` - massiv tartiblangan qidirilayotgan elementning indeksini qaytaradi

```python
a = np.array([3, 17, 5, 6, 4, 12, 3]) #3 3 4 5 6 12 17

a.searchsorted(6)
```

`sort()` - massivni saralash

`unique()` - massiv elementlarini takrorlamagan holda qaytaradi

```python
a = np.array([3, 4, 3, 4, 0, -1, 0])

frozenset(np.unique(a))
```

`count_nonzero()` - massivning qiymati nolga teng bo'lmagan elementlari soni

```python
a = np.array([3, 4, 3, 4, 0, -1, 0])

np.count_nonzero(a)
```

`argmax()` - massivdagi eng katta elementning indeksini qaytaradi

```python
a = np.array([3, 4, 3, 4, 0, -1, 0])

a.argmax()

a.max()
```

`argmin()` - massivdagi eng kichik elementning indeksini qaytaradi

```python
a.argmin()

a.min()
```

`repeat(n)` - massiv elementlarining har biri n martadan takrorlangan yangi massiv qaytaradi

```python
array1 = a.repeat(3)
```

`full((m, n), value)` - massivni berilgan qiymat(lar) bilan berilgan o'lchamda to'ldirish

```python
array2 = np.full((3, 3), 0)
```

# **Fayllar bilan ishlash**

`save()` - massivni fayl shaklida saqlash

```python
np.save('ahmoqona_massiv', a)
```

`load()` - massiv(lar) saqlangan mavjud faylni yuklash

```python
array = np.load('ahmoqona_massiv.npy')

array
```

`savez()` - bir nechta massivni bitta faylga saqlash

```python
np.savez('ikkita_ahmoqona_massiv', a=array1, b=array2) #kalit berib saqlanadi massivlar
```

```python
arrays = np.load('ikkita_ahmoqona_massiv.npz')

arrays['a'] #kalitlar orqali o'qib olinadi

arrays['b']
```

`savez_compressed()` - massivlarni siqilgan holatda faylga saqlash

```python
np.savez_compressed('siqilgan_ahmoqona_massivlar', a=array1, b=array2)
```

```python
arrays = np.load('siqilgan_ahmoqona_massivlar.npz')

arrays['b']
```
