#**Yagona standart qiymat bilan to'ldirish**

```python
import scipy.stats as st

gender = df['driver_gender']

my_dict = {
    'M': 0,
    'F': 1,
    'MF': 2,
    'W': 3
}

binary_gender = gender.map(my_dict)

list(st.mode(binary_gender))[0]
```
