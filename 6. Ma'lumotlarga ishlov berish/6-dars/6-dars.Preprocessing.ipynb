# **Ma'lumotlar to'plamlarini birlashtirish**

### `Merging`

```python
import pandas as pd
```

```python
people = {
    'personId': [1, 2],
    'lastName': ['Wang', 'Alice'],
    'firstName': ['Allen', 'Bob']
}
```

```python
addresses = {
    'addressId': [1, 2],
    'personId': [2, 3],
    'city': ['New York City', 'Leetcode'],
    'state': ['New York', 'California']
}
```

```python
people = pd.DataFrame(people)
address= pd.DataFrame(addresses)
```

```python
people
```

```python
address
```

```python
people.merge(address)
```

```python
people.merge(address, how='outer')
```

```python
people.merge(address, how='left')
```

```python
people.merge(address, how='right')
```

```python
people.merge(address, how='inner')
```

### `concatenate`

```python
pd.concat([people, address], axis=0)
```

```python
pd.concat([people, address], axis=1)
```
