# **Data Vizualization (Vizuallashtirish)**

### **with Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#grafik chiziladigan maydonni hosil qilish
plt.figure(figsize=(13, 6))


#x va y qiymatlarni hisoblash
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)


#sin(x) funksiyani chizish
plt.plot(x, sin,
         linewidth=7,
         linestyle='--',
         color='#504B8E')


#cos(x) funksiyani chizish
plt.plot(x, cos,
         linewidth=3,
         linestyle=':',
         color='y')


#setka (to'r) qo'shamiz
plt.grid() 

#x o'qidagi sonlarga izoh yozamiz
plt.xlabel('x -> [0; 4pi] sonlar')

#y o'qidagi sonlarga izoh yozamiz
plt.ylabel('sin(x) qiymatlar')

#grafik uchun izoh yozamiz
plt.suptitle('sin(x) va cos(x) funksiyalarning grafigi')

#tupoylarga chizmalar nimani ifodalagini ko'rsatish uchun
plt.legend(['sin(x)', 'cos(x)'], loc=3)

#chizmaga matn qo'shish 
plt.text(x=2.4, y=0.8, s='Buni men chizdim', fontdict={'fontsize': 14, 'fontweight': 'bold'})

#barcha hosil qilinganlarni ekranga chiqarish
plt.show()
```

### **with Seaborn**

### **Tendensiyani tasvirlash**

```python
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/iRavshan/pro_data_scientist/main/datasets/poverty.csv')
```

```python
df.head()
```

```python
df['Year'].describe()
```

```python
df['Name'].unique()
```

```python
pov = df[df['Name']=='Washington']
```

```python
pov.drop(['ID', 'Name', 'Poverty Universe', '90% Confidence Interval', 'Percent in Poverty', '90% Confidence Interval.1'], axis=1, inplace=True)

pov.set_index('Year', inplace=True)
```

```python
pov
```

```python
plt.figure(figsize=(13, 8))

sns.lineplot(data=pov, markers='o')

plt.xlabel('Yillar')

plt.ylabel("Ko'rsatkichlar")

plt.title("Washington shtatining qashshoqlik ko'rsatkichi", pad=10)

plt.grid()

plt.annotate('Maksimum qashshoqlik', xy=(2013, 960000), xytext=(2013, 968000))

plt.annotate('Minimum qashshoqlik', xy=(2020, 714653), xytext=(2020, 705000))

plt.savefig('Washington.png')

plt.show()
```
