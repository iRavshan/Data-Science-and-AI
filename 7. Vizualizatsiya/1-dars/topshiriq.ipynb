```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(20, 8))

x = np.arange(-4 * np.pi, 4 * np.pi, 0.01)
sin = np.sin(x)
cos = np.cos(x)

plt.plot(x, sin,
         linewidth=5,
         linestyle='-.',
         color='#50C878')

plt.plot(x, cos,
         linewidth=10,
         linestyle='--',
         color='#504B8E')

plt.grid()

plt.xlabel('x  [-4; 4]',
           fontdict={'size': 16, 'weight': 'semibold', 'family': 'Serif'},
           labelpad=8)

plt.ylabel('funksiyalar qiymatlari',
           fontdict={'size': 16, 'weight': 'semibold', 'family': 'Serif'},
           labelpad=8)

plt.title('sin(x) va cos(x) funksiyalar grafiklari',
          fontdict={'size': 20, 'weight': 'bold', 'family': 'Serif', 'color': '#263238'},
          loc='center',
          pad=20)

plt.legend(['sin(x)', 'cos(x)'], loc=3)

plt.savefig('grafik.png')

plt.show()
```
